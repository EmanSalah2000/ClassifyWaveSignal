```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
```

```python

def plot_signal(samplerate, signal):
    length = len(signal) / samplerate
    time = np.linspace(0., length, len(signal))
    plt.plot(time, signal, label="signal")
    plt.legend()
    plt.xlabel("Time [s]")
    plt.ylabel("Amplitude")
    plt.show()
```

```python
def plot_frames(frames):
    plt.figure(figsize=(20, 4))
    for i in range(10):
        plt.subplot(2, 5, i + 1)
        plt.plot(frames[i], label="frame")
    plt.show()
```

```python
def framing(name_file, frame_size, overlab_size):
    frames = []
    samplerate, data = wavfile.read(name_file)
    number_samples_per_frame = int(frame_size * samplerate)
    number_samples_in_overlab = int(overlab_size * samplerate)
    step = (number_samples_per_frame - number_samples_in_overlab)

    for i in range(0, len(data), step):
        frame = np.zeros([number_samples_per_frame, ])
        frame[0:len(data[i:i + number_samples_per_frame])] = data[i:i + number_samples_per_frame]
        frames.append(frame)

    print("\nnumber of samples per frame: ", number_samples_per_frame)
    print("Len of frames: ", len(frames), "\n")

    return data ,samplerate,frames
```

```python
frame_size = 0.020
overlab_size = 0.010
name_file='Welcome.wav'
data,samplerate,frames = framing(name_file, frame_size, overlab_size,)

print("samplerate: ", samplerate)
print("Data shape: ", data.shape, "\n")
plot_signal(samplerate, data)
print("frame 0: ", frames[0], "\n")
print("frame ", len(frames) - 1, " : ", frames[len(frames) - 1])
# print("frame ", "19", " : ", frames[20])

```

```python
plot_frames(frames)
```

```python
signal = []
for frame in frames:
    for item in frame:
        signal.append(item)

plot_signal(samplerate, signal)
```
